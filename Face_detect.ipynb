```python
import cv2
```

```python
img = cv2.imread(r"C:\Users\aswat\Desktop\DL PROJECTS\Face detection\[Photos] New Behind the Scenes Images Added for the Korean Drama 'The Penthouse'.jpg")

img
```

```python
import matplotlib.pyplot as plt
plt.imshow(img)
```

```python
converted_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(converted_img)
```

```python
face_cascade = cv2.CascadeClassifier(r'C:\Users\aswat\Desktop\DL PROJECTS\Face detection\haarcascade_frontalface_default.xml')
face_cascade
```

```python
img_gray=cv2.cvtColor(converted_img,cv2.COLOR_RGB2GRAY)
img_gray
```

```python
faces=face_cascade.detectMultiScale(img_gray,minNeighbors=15)
faces
```

```python
len(faces)
```

```python
for(x,y,w,h) in faces:
    cv2.rectangle(converted_img,(x,y),(x+w,y+h),(0,0,255),6)
```

```python
video=cv2.VideoCapture(r'C:\Users\aswat\Desktop\DL PROJECTS\Face detection\WhatsApp Video 2023-11-15 at 11.36.52 AM.mp4')
while True:
    success,frame=video.read()
    gray_img=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    faces=face_cascade.detectMultiScale(gray_img,minNeighbors=14)
    # print(len(faces))
    for (x,y,w,h) in faces:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,255),4)
    cv2.imshow('Video',frame)
    if cv2.waitKey(1) & 0XFF==ord('q'):
        break
# video.release()
cv2.destroyAllWindows()
```

```python
import cv2

video = cv2.VideoCapture(0)

while True:
    success, frame = video.read()
    grayimg=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    faces=face_cascade.detectMultiScale(grayimg,minNeighbors=20)
    for(x,y,w,h) in faces:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,255),4)
        
    
    cv2.imshow('Webcam', frame)
    
    if cv2.waitKey(1) & 0xFF == ord('a'):
        break

video.release()
cv2.destroyAllWindows()
```
