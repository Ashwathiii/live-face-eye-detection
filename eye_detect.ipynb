```python
import cv2
```

```python
img = cv2.imread(r"C:\Users\aswat\Desktop\DL PROJECTS\Face detection\[Photos] New Behind the Scenes Images Added for the Korean Drama 'The Penthouse'.jpg")

img
```

```python
import matplotlib.pyplot as plt
plt.imshow(img)
```

```python
converted_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(converted_img)
```

```python
eye_cascade=cv2.CascadeClassifier(r"C:\Users\aswat\Desktop\DL PROJECTS\Face detection\haarcascade_eye.xml")
eye_cascade
```

```python
img_gray=cv2.cvtColor(converted_img,cv2.COLOR_RGB2GRAY)
img_gray
```

```python
eyes=eye_cascade.detectMultiScale(img_gray,minNeighbors=6)
eyes
```

```python
len(eyes)
```

```python
for (x,y,w,h) in eyes:
    cv2.rectangle(converted_img,(x,y),(x+w,y+h),(0,0,255),4)
```
